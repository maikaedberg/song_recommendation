```python
checked = dict()
```

```python
def filter_non_verified(checked):
    verified_count = 0
    total_checked = 0
    file = open('unique_artists.txt', 'r')
    for line in file:
        l = line.strip().split('<SEP>')
        artist = l[3]
        try:
            if artist[0] == '-':
                artist = artist[1:]
            if artist == "!!!":
                continue
            if artist not in checked:
                search_res = api.search_users(artist, count = 1, wait_on_rate_limit=True)
                found = 0
                for user in search_res:
                    if user.verified:
                        checked[artist] = user.id
                        verified_count += 1
                        found = 1
                        break
                if found == 0:
                    checked[artist] = -1
            else:
                #print("Artist checked")
                pass
            total_checked += 1
            if total_checked % 100 == 0:
                print("Checkpoint: " + str(total_checked))
            
        except tweepy.error.RateLimitError:
            print("Checked " + str(total_checked) + " lines, verified = " + str(verified_count))
            file.close()
            return checked

    file.close()
    return checked

def line_n(filename):
    count = 0
    file = open(filename, 'r')
    for line in file:
        count += 1
    file.close()
    return count
    

def different_artists(filename):
    checked2 = dict()
    file = open(filename, 'r')
    for line in file:
        l = line.strip().split('<SEP>')
        artist = l[3]
        if artist not in checked2:
            checked2[artist] = 1
    file.close()
    return len(checked2)
    
filter_non_verified(checked)
print(len(checked))
print(different_artists('unique_artists.txt'))
```

```python
def update_ids():
    file = open("twitter_ids.txt", "w")
    for artist in checked:
        if checked[artist] != -1:
            file.write(artist + "<SEP>" + str(checked[artist]) + "\n")
    file.close()
    
    file = open("checked.txt", "w")
    for artist in checked:
        file.write(artist + "<SEP>" + str(checked[artist]) + "\n")
    file.close()
    
update_ids()

print("Found ids = " + str(line_n("tweeter_ids.txt")))
print("Total artists checked " + str(line_n("checked.txt")))
```

```python
def compare_artfile():
    counter2 = 0
    checked3 = dict()
    file = open("unique_artists.txt", 'r')
    for line in file:
        artist = line.strip().split('<SEP>')[3]
        if artist in checked:
            if artist not in checked3:
                counter2 += 1
                checked3[artist] = 1
    file.close()
    return(counter2)
    
compare_artfile()
```
